```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../out_matrix.csv")
df.fillna(0, inplace=True)
df = df.drop("sample", axis=1)
```

```python

df.set_index("Unnamed: 0", inplace=True)
df = df.astype(float)
df
```

```python
from sklearn.decomposition import PCA

# Create a PCA instance
pca = PCA(n_components=2)  # Specify the number of components you want to keep

# Apply PCA to the DataFrame
pca_result = pca.fit_transform(df.T.values)

# Create a new DataFrame with the PCA results
df_pca = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])

# Display the DataFrame with PCA results
print(df_pca)
```

```python
plt.scatter(df_pca['PC1'], df_pca['PC2'])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Scatter Plot')
plt.show()
```
